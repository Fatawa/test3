```python
import requests
import csv
```

```python
response = requests.get("https://www.metaweather.com/api/location/1521894/")
response
```

```python
weather = response.json()
weather
```

```python
days = weather['consolidated_weather']
days
```

```python
days = days[0:3]
days
```

```python
for i in days:
    print(i['the_temp'], "|", i['humidity'], "|", i['visibility'], "|", i['air_pressure'], "|", i['wind_speed'], "|", i['applicable_date'])
```

```python
with open("weather.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames= ['temp', 'humidity', 'visibility' , 'air_pressure', 'wind_speed', 'date_time'])
    writer.writeheader()
    for i in days:
        temp = i['the_temp']
        humidity = i['humidity']
        visibility = i['visibility']
        air = i['air_pressure']
        wind = i['wind_speed']
        date = i['applicable_date']
        writer.writerow(
            {
                'temp' : temp,
                'humidity' : humidity,
                'visibility' : visibility,
                'air_pressure' : air,
                'wind_speed' : wind,
                'date_time' : date
            }
        )

```
